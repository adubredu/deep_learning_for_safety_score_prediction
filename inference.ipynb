```python
from fastai.vision import *
from fastai.metrics import error_rate
```

```python
data=ImageList.from_folder('data/stamford_images_only')
```

```python
learner = load_learner('model/', test=data)
```

```python
hart_preds,y = learner.get_preds(ds_type=DatasetType.Test)
```

```python
len(hart_preds)
```

```python
names = data.items
print(names[0])
print(str(names[1].stem).split('/'))
```

```python
names = data.items
image_names=[]
for name in names:
    parts = str(name.stem).split("/")
    image_names.append(parts[0]+'.jpg')
```

```python
SCORES
```

```python
p = hart_preds.numpy()
import numpy as np
indexes = np.argmax(p, axis=1)

SCORES=[]
for i in indexes:
    SCORES.append(i)
```

```python
import pandas as pd
hdf = pd.DataFrame(image_names)
hdf.columns=['filename']
hdf['scores'] = SCORES

hdf.to_csv('stamford_predictions.csv', index=False)
```
